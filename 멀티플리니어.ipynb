```python
import pandas as pd

# 데이터 불러오기
data = pd.read_csv('5.csv')

# 수익률 계산
data['수익률'] = data['미국'].pct_change(fill_method=None)

# 새로운 열 생성
data['미국_1'] = data['수익률'].apply(lambda x: 1 if x > 0 else 0)
data['미국_0'] = data['수익률'].apply(lambda x: 1 if x < 0 else 0)
data=data.dropna()
data = data.iloc[:, [0, 1, 3, 4, 5, 2]]

# 두 개의 열 제거
data.drop(data.columns[[1, 2]], axis=1, inplace=True)
```
