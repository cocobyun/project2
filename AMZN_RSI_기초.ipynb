```python
import finterstellar as fs
```

```python
df=fs.get_price('AMZN',start_date='2024-01-01', end_date='2024-10-31')
df
fs.draw_chart(df,right='AMZN')
```

```python
fs.rsi(df,w=14)
fs.draw_chart(df, left='rsi', right='AMZN')
```

```python
fs.indicator_to_signal(df,factor='rsi',buy=40,sell=60)
fs.position(df)
fs.draw_chart(df,left='rsi', right='position_chart')
```

```python
fs.evaluate(df, cost=0.001)
fs.draw_chart(df, left='acc_rtn_dp', right='AMZN')
```

```python
fs.performance(df, rf_rate=0.01)
```
